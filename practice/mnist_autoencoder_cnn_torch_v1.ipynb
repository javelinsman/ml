```python
from mnist_autoencoder_cnn_torch_v1 import fetch_dataloaders, Autoencoder, train, test

train_loader, test_loader = fetch_dataloaders()
```

```python
model = Autoencoder()
model
```

```python
train(model, train_loader, epochs=50)
```

```python
import torch
import matplotlib.pyplot as plt
import numpy as np
import torchvision

def show(images_tensor):
    arrs = list(images_tensor)
    plt.figure(figsize=(15, 5))
    plt.imshow(np.transpose(torchvision.utils.make_grid(arrs), (1, 2, 0)))
    plt.show()

for i, (images, _) in enumerate(test_loader):
    if i >= 10:
        break
    show(images)
    show(test(model, images))
    print('\n\n')
```
